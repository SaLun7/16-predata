```python
# 나눗셈 후 소수점 이하를 버리는 // 연산자
print(5 // 2)
```

```python
# 나눗셈 나머지를 구하는 % 연산자
print(5 % 2)
```

```python
# 거듭제곱을 구하는 ** 연산자
print(2 ** 10)
```

```python
# 값을 정수로 만들기
a = 3.3
print(int(a))

b = '10'
print(type(b))
b = int(b)
print(type(b))
```

```python
# 실수 계산하기
print(3.5 + 2.1)
```

```python
# 실수와 정수를 함께 계산하기
print(4.2 + 5)
```

```python
# float(숫자)
print(float(5))
print(float(1 + 2))
print(float('5.3'))

a = float('5.3')
print(type(a))
```

```python
# 괄호 사용하기
print((35 + 1) * 2)
```
