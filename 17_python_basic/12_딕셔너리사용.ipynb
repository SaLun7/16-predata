```python
# 딕셔너리 만들기

# 딕셔너리 = {
#     키1: 값1,
#     키2: 값2,
#     ...
# }

lux = {
    "health" : 490,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}
```

```python
# 키 이름이 중복되면?

lux = {
    "health" : 490,
    "health" : 800,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}

print(lux)
```

```python
# 딕셔너리 키의 자료형
x = {
    100 : "hundred",
    False : 0,
    3.5 : [3.5, 3.5]
}

print(x)
```

```python
x = {
    [10, 20] : 100
}

print(x)
```

```python
# 빈 딕셔너리 만들기

x = {}
y = dict()

print(x)
print(y)

print(type(x))
print(type(y))
```

```python
# dict로 딕셔너리 만들기

lux1 = dict(
    health = 490,
    mana = 334,
    melee = 550,
    armor = 18.72
)

print(lux1)
```

```python
lux2 = dict(zip(
    ["health", "mana", "melee", "armor"],
    [490, 334, 550, 18.72]
))

print(lux2)
```

```python
lux3 = dict([
    ("health", 490),
    ("mana", 334),
    ("melee", 550),
    ("armor", 18.72)
])

print(lux3)
```

```python
lux4 = dict({'health': 490, 'mana': 334, 'melee': 550, 'armor': 18.72})
print(lux4)
```

```python
# 딕셔너리 키에 접근하고 값 할당하기

lux = {'health': 490, 'mana': 334, 'melee': 550, 'armor': 18.72}

print(lux["health"])
print(lux["armor"])
```

```python
lux["health"] = 2037
lux["mana"] = 1104

print(lux)
```

```python
lux["mana_regen"] = 3.28
print(lux)
```

```python
print(lux["attack_speed"])
```

```python
print(lux.get("attack_speed"))
```

```python
# 딕셔너리에 키가 있는지 확인하기
print("health" in lux)
print(2037 in lux)
```

```python
#딕셔너리 키 개수 구하기
print(len(lux))
```
